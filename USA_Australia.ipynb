## Data Cleaning and Conversion 

<br>

It is import to compare two dataset's structure and content, in order to decide an approach how the data should be cleaned and organised. 
After the preliminary inspection above. The US data set has been chosen to be cleaned first.
Comments have been made in the coding cells below:

```python
# assign our two csv files to two constants.

US_DATA = "ACSST5Y2020.S1201-2022-08-25T015155.csv"
AU_DATA = "2021Census_G05_AUS_AUS.csv"
```

```python
def cleaner (data_row):
    '''a function to remove with all the quotation marks in the US data, also remove all the commas in numbers.'''
    data_row = data_row.strip()
    while '"' in data_row:
        first_index = data_row.find('"')
        second_index = data_row.index('"',first_index+1)
        new_string = data_row [first_index:second_index+1].replace('"','').replace(',','')
        data_row = data_row.replace (data_row[first_index:second_index+1], new_string)
    return(data_row)
```

<br>

### US data cleansing : 
<br>

```python
# read in files, drop the unnecessry rows.then use slice object to keep the wanted columns. 

# Delete all the \xa0 Unicode represents a hard space or a no-break space. 

# divide all the percentages by 100, to represent the ratio for better calculations. 4 decimal points, remove percentage signs.

with open (US_DATA, "r", encoding='utf-8') as us_data :
    raw_us_list = us_data.readlines()
    raw_us_list = raw_us_list[:17]
    us_list = []
    wanted = slice(1,None,2)
    for line in raw_us_list:
        list_header = []
        line = cleaner (line)
        line = line.replace("\xa0","")
        my_list = line.split(',')
        list_header.append (my_list[0])
        my_list = list_header + my_list[wanted]
        for idx in range(len(my_list)):
            if "%" in my_list[idx]:
                my_list[idx] = round(float(my_list[idx].strip("%"))/100,4)
        us_list.append(my_list)


print(us_list[:6])
```

```python
# furthur remove unwanted lines, rename the category names.

# re_order the row order, to be the same with australian list: Total-Count _ Married _ Separated _ Divorced _ Widowed _ Never Married. 

# refomat the group names, to be more clear and readable, to be consistant with Australian Census data.

us_list.pop(2)
us_label = us_list.pop(0)

for item in us_list:
    item[1] = int(item[1])
    item[3],item[5] = item[5],item[3]
    item[0] = item[0].replace(" to ","_").replace(" years","_yr")
    
    
    
us_list [0][0] = "P_Tot"
us_list [1][0] = "M_Tot"
us_list [7][0] = "65_over"
us_list [8][0] = "F_Tot"
us_list [-1][0] = "65_over"

for item in us_list[2:8]:
    item[0] = 'M_'+item[0]
for item in us_list[9:]:
    item[0] = 'F_'+item[0]
us_list

```

```python
# Transform all the percentage to population numbers, by multiple total population in each group and ratio

ratio_to_pop = []

for item in us_list:
    ratio_to_pop.append(item[:2] + [int(item[1]*i) for i in item[2:]])

ratio_to_pop
```

```python
# US cencus data didn't include Male+Female marital status for each age bracket. The above generated population list will be used here to produce a "person_list" for each age bracket.

# Add all the F and M for each age group and each marital status category, then transform into Percentage

person_list = []

for idx in range(2, int((len(ratio_to_pop)+1)/2)):
    temp_list = []
    temp_list.append(ratio_to_pop[idx][0].replace("M","P",1))
    temp_list.append(ratio_to_pop[idx][1]+ratio_to_pop[idx+7][1])
    for i in range(2,len(ratio_to_pop[idx])):
                   temp_list.append (round(((ratio_to_pop[idx][i]+ratio_to_pop[idx+7][i])/temp_list[1]),4))
    person_list.append(temp_list)

person_list
```

```python
# combine person_list with us_list. move the first item of "Tot" to the last
us_list += person_list
us_list.append(us_list.pop(0))
us_list
```

### Australian Data Cleaning 

<br>

```python
# Read in Australia data file.
# split, and group every 6 columns, to represent each "gender_age" bracket. Do the same for both header row and data row.

with open (AU_DATA, "r", encoding='utf-8') as au_data :
    raw_au_list = []
    for line in au_data:
        my_list = line.strip().split(",")
        my_list = my_list[1:]
        list_chunked = [my_list[i:i + 6] for i in range(0, len(my_list), 6)]
        raw_au_list.append(list_chunked)

print(raw_au_list[0][:4])
print(raw_au_list[1][:4])
```

```python
# Transform and format of the data:

# 1: Retrive each header from hearder row.

# 2: Retrive total population count from data row 

# 3: Calculate and replace all the population count into ratio, to 4 decimal points

au_list = []

for idx in range(len(raw_au_list[0])):
    temp_list=[]
    temp_list.append(raw_au_list[0][idx][0].rpartition("_")[0])
    temp_list.append(int(raw_au_list[1][idx][-1]))
    temp_list = temp_list + raw_au_list[1][idx][:-1]
    for i in range(2,len(temp_list)):
        temp_list[i] = round( (int(temp_list[i])/temp_list[1]) , 4)
    au_list.append(temp_list)

au_list [:5]
    
```

```python
# The US data has combined 20 - 34 years old into one age group. and there is no sub groups above 65 years and over

# In order to have more meaning comparison. Two new list " age 20-34 " and " 65 years and over " are created for the australian data


add_on_list = []
M_age_20_34 = ['M_20_34_yr'] + [int(x) + int(y) for x, y in zip(raw_au_list[1][1], raw_au_list[1][2])]
add_on_list.append(M_age_20_34)
F_age_20_34 = ['F_20_34_yr'] + [int(x) + int(y) for x, y in zip(raw_au_list[1][11], raw_au_list[1][12])]
add_on_list.append(F_age_20_34)
M_65_over = ['M_65_over'] + [int(x) + int(y) + int(z) for x, y, z in zip(raw_au_list[1][6], raw_au_list[1][7], raw_au_list[1][8])]
add_on_list.append(M_65_over)
F_65_over = ['F_65_over'] + [int(x) + int(y) + int(z) for x, y, z in zip(raw_au_list[1][16], raw_au_list[1][17], raw_au_list[1][18])]
add_on_list.append(F_65_over)
P_age_20_34 = ['P_20_34_yr'] + [int(x) + int(y) for x, y in zip(raw_au_list[1][21], raw_au_list[1][22])]
add_on_list.append(P_age_20_34)
P_65_over = ['P_65_over'] + [int(x) + int(y) + int(z) for x, y, z in zip(raw_au_list[1][-2], raw_au_list[1][-3], raw_au_list[1][-4])]
add_on_list.append(P_65_over)

for item in add_on_list:
    tot = item.pop(-1)
    item.insert(1,tot)

for item in add_on_list:
    for idx in range(2,len(item)):
        item[idx] =  round( item[idx]/item[1] , 4)
add_on_list
        
```

```python
# Append add_on_list to Au_list

au_list += add_on_list
au_list.sort()

au_list [:5]
```

```python
# Now there are two very similar and comparable data for further investigating the relationships

# Adding a title row to both list

title_row = ["Category","Total", "Married" , "Separated" , "Divorced" , "Widowed" , "Never Married" ] 
us_list.insert(0,title_row)
print(us_list[:5])
```

```python
au_list.insert(0,title_row)
print(au_list[:5])
```

```python
# Lastly, covert both lists into dictionaries.

au_dict = {}
us_dict = {}
for item in au_list:
    au_dict[item[0]] = item[1:]
for item in us_list:
    us_dict[item[0]] = item[1:]
```
